```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("C:\\Users\\Arun N\\Downloads\\customer_locations.csv") 
X = data[['Latitude', 'Longitude']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
db = DBSCAN(eps=0.5, min_samples=10)
db.fit(X_scaled)
```

```python
data['Cluster']= db.labels_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['Latitude'], data['Longitude'], c=data['Cluster'], cmap='rainbow')
plt.title("DBSCAN Clustering of Customer Locations")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.cluster import KMeans
Kmeans = KMeans(n_clusters=3)
Kmeans.fit(X)
print(Kmeans.labels_)
print(Kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(*zip(*X.values),c=Kmeans.labels_,cmap='rainbow',marker=1)
plt.scatter(*zip(*Kmeans.cluster_centers_),color='red',marker='x',s=200)
plt.title('kmeans clustering of customer locations')
plt.xlabel('logitude')
plt.ylabel('latitude')
plt.show()
```
