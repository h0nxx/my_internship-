```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


```

```python
df = pd.read_csv("C:\\Users\\Arun N\\Downloads\\Titanic-Dataset.csv")
```

```python
df
```

```python
df = df[['Pclass','Sex','Age','Survived']]
df.dropna(inplace=True)
df['Sex'] = df['Sex'].map({'male':0,'female':1})
```

```python
x = df[['Pclass','Sex','Age']]
y = df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

model = DecisionTreeClassifier(criterion='gini',max_depth=3)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(accuracy_score (y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plot_tree(model,feature_names=x.columns,class_names=['Not Survived','Survived'],filled=True)
plt.show()
```

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
#select features
data = df[['pclass','sex','age','sibsp','parch','fare','embarked','survived']]

#handle missing values
data.dropna(inplace=True)

#convert  categories to numeric
data['sex'] = data['sex'].map({'male':0,'female':1})
data['embarked']=data['embarked'].map({'S':0,'C':1,'Q':2})

#split x n y
x = data.drop('survived',axis=1)
y = data['survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

#Accuracy
print("Accuracy:",accuracy_score(y_test,y_pred))

#Confusion Matrix
print(confusion_matrix(y_test,y_pred))

#classification Report
print(classification_report(y_test,y_pred))
```

```python
importances = model.feature_importances_
feature_names = x.columns
feature__imp_df = pd.DataFrame({'Feature': feature_names,'Importance':importances})
feature__imp_df.sort_values(by='Importance',ascending=False,inplace=True)

sns.barplot(x='Importance',y='Feature',data=feature__imp_df)
plt.title("feature Importance")
plt.show()
                               
```
