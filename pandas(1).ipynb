```python
sns.histplot(df['total_bill'],kde=True,color='blue')
plt.title("Distribution of total_bill")
plt.xlabel("total_bill")
plt.ylabel("count")
plt.show()
```

```python
sns.boxplot(x ='day',y ='total_bill',data = df,palette='Set1')
plt.title("Total Bill by Day")
plt.show()


```

```python
sns.boxplot(x ='sex',y ='tip',data = df,palette='Set1')
plt.title("Average tip by gender")
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=df,)
plt.title("tip vs total_bill")
plt.show()
```

```python
sns.barplot(x = 'sex',y = 'tip',data = df)
plt.title("Average tip by gender")
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=df)
plt.title("total_bill  vs tip")
plt.show()
```
